```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# loading dataset
data = sns.load_dataset("iris")

# draw lineplot
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.title('Title using Matplotlib Function')
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
plt.xlim(5)

plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# changing the theme to dark
sns.set_style("dark")
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Removing the spines
sns.despine()
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# Setting the scale of the plot
sns.set_context("paper")

plt.show()
```

```python
def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

with sns.axes_style('darkgrid'):

    # Adding the subplot
    plt.subplot(211)
    plot()

plt.subplot(212)
plot()
```

```python
palette = sns.color_palette()


sns.palplot(palette)

plt.show()
```

```python
palette = sns.color_palette('PiYG', 11)


sns.palplot(palette)

plt.show()
```

```python
def plot():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

sns.set_palette('vlag')
plt.subplot(211)


plot()


sns.set_palette('Accent')
plt.subplot(212)
plot()

plt.show()
```

```python
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)


fig = plt.figure(figsize =(5, 4))

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

graph()

ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])

graph()

plt.show()
```

```python
def graph():
    sns.lineplot(x="sepal_length", y="sepal_width", data=data)

axes1 = plt.subplot2grid ( (7, 1), (0, 0), rowspan = 2,  colspan = 1)
graph()

axes2 = plt.subplot2grid ( (7, 1), (2, 0), rowspan = 2, colspan = 1)
graph()

axes3 = plt.subplot2grid (
  (7, 1), (4, 0), rowspan = 2, colspan = 1)
graph()
```
